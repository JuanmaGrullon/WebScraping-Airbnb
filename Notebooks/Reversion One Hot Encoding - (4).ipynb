```python
import pandas as pd
import numpy as np
```

```python
# Cargar el DataFrame imputado
df_imputed = pd.read_excel('../data_out/Hospedajes_RD-IMPUTED V2.xlsx')
```

```python
df_imputed.info()
```

```python
# Definir las columnas numéricas y categóricas
numerical_columns = ['ID_ALOJAMIENTO','PRECIO_USD', 'MAX_HUESPEDES', 'CANTIDAD_HABITACIONES', 'CANTIDAD_CAMAS', 'CANTIDAD_BAÑOS',
                     'ES_SUPER_ANFITRION', 'CANTIDAD_EVALUACIONES', 'CALIFICACION_GENERAL', 'CALIFICACION_LIMPIEZA',
                     'CALIFICACION_VERACIDAD', 'CALIFICACION_LLEGADA', 'CALIFICACION_COMUNICACION',
                     'CALIFICACION_UBICACION', 'CALIFICACION_CALIDAD_PRECIO','FLAG_RESERV_INMEDIATA','FLAG_LLEGADA_AUTONOMA','FLAG_CANCELAC_GRATUITA','FLAG_ACEPT_MASCOTAS']

categorical_columns = ['TIPO_ALOJAMIENTO', 'ESPACIO_ALOJAMIENTO', 'UBICACION']
```

```python
# Invertir el proceso de OneHotEncoding
def revert_one_hot_encoding(df, categorical_columns):
    df_reverted = df.copy()
    
    for col in categorical_columns:
        # Encontrar las columnas que pertenecen a la variable categórica original
        one_hot_cols = [c for c in df.columns if c.startswith(col + '_')]
        
        # Obtener el nombre original de la variable categórica
        def get_original_category(row):
            for c in one_hot_cols:
                if row[c] == 1:
                    return c.replace(col + '_', '')
            return np.nan
        
        # Aplicar la función para revertir OneHotEncoding
        df_reverted[col] = df.apply(get_original_category, axis=1)
        
        # Eliminar las columnas OneHotEncoded
        df_reverted.drop(columns=one_hot_cols, inplace=True)
    
    return df_reverted
```

```python
# Revertir las variables categóricas
df_reverted = revert_one_hot_encoding(df_imputed, categorical_columns)
```

```python
# Guardar el DataFrame resultante
df_reverted.to_excel('../data_out/Hospedajes_RD- V PBI.xlsx', index=False)
```
